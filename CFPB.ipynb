```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 6))
colors = sns.color_palette('pastel')[0:2]  # Using Seaborn color palette
df['Timely response?'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=colors)

# Set a title using Seaborn style
plt.title('Distribution of Timely Responses', fontsize=16)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

sns.set(style="whitegrid")

# Using countplot to visualize the count of occurrences for 'Company response to consumer' column
ax = sns.countplot(x='Company response to consumer', data=df, palette='viridis')

plt.title('Distribution of Company Responses to Consumers')
plt.xlabel('Company Response to Consumer')
plt.ylabel('Count')

# Add percentages on top of the bars
total = len(df['Company response to consumer'])
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2.,
            height + 3,
            f'{height/total:.1%}',
            ha="center")

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

```python
#Population density
df2 = pd.read_csv("/kaggle/input/population-data/pop data.csv")
```

```python
import pandas as pd

# Assuming population_df is your population dataset
# Assuming complaints_df is your DataFrame with complaints

# Merge the datasets on the 'State' column
merged_df = pd.merge(df2, df.groupby('State').size().reset_index(name='Number of Complaints'), on='State', how='left')

# Fill NaN values with 0 for states with no complaints
merged_df['Number of Complaints'].fillna(0, inplace=True)

# Display the resulting DataFrame
print(merged_df)
```

```python
merged_df.drop(merged_df[merged_df['State'] == 'DC'].index, inplace=True)
```

```python
print(merged_df)
```

```python
merged_df.dropna(axis=0,inplace=True)
```

```python
import pandas as pd

# Assuming state_area_data is a list containing the state area data
state_area_data = [135767, 665384, 113990, 53179, 163695, 104094, 5543, 2489, 65758, 59425,
                   10932, 83569, 57914, 36420, 56273, 82278, 40408, 52378, 35380, 12406,
                   10554, 96714, 86936, 48432, 69707, 147040, 77348, 110572, 9349, 8723,
                   121590, 54555, 53819, 70698, 44826, 69899, 98379, 46054, 1545, 32020,
                   77116, 42144, 268596, 84897, 9616, 42775, 71298, 24230, 65496, 97813, 5325]

# Create a DataFrame with 'State' and 'Area' columns
area_df = pd.DataFrame({
    'State': merged_df['State'],  # Assuming 'State' column already exists in merged_df
    'Area(Sq Mi)': state_area_data
})

# Display the resulting DataFrame
print(area_df)
```

```python
# Assuming merged_df is your existing DataFrame
# Assuming area_df is the DataFrame containing state area data

# Merge the DataFrames on the 'State' column
merged_df = pd.merge(merged_df, area_df, on='State', how='left')

# Display the resulting DataFrame
print(merged_df)
```

```python
merged_df['Population'] = pd.to_numeric(merged_df['Population'], errors='coerce').astype('Int64')
```
